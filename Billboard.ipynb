```python
def collect_song_names():
    popular_songs = []
    unpopular_songs = []

    chart = billboard.ChartData('hot-100')  # Fetch the Hot 100 chart
    for song in chart:
        # Check if the song has been on the chart for more than 10 weeks
        if song.weeks > 10:
            popular_songs.append((song.title, song.artist))
        else:
            unpopular_songs.append((song.title, song.artist))

    return popular_songs, unpopular_songs

# Call the function to collect song names
popular_songs, unpopular_songs = collect_song_names()

# Print the results including artist names
print("Popular Songs:")
for title, artist in popular_songs:
    print(f"{title} by {artist}")

print("\nUnpopular Songs:")
for title, artist in unpopular_songs:
    print(f"{title} by {artist}")
```

```python
import youtube_dl
import logging

# Configure logging
logging.basicConfig(level=logging.INFO)

# Merge popular and unpopular songs along with their respective artists
all_songs = popular_songs + unpopular_songs

# Initialize youtube_dl downloader
ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'outtmpl': 'downloaded_songs/%(title)s_%(artist)s.%(ext)s',
    'default_search': 'ytsearch',
}

# Batch download songs
with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    for song, artist in all_songs:
        # Construct search query
        search_query = f'{song} {artist} official audio'

        try:
            # Download the first search result
            ydl.download([search_query])
            logging.info(f"Downloaded: {song} by {artist}")
        except Exception as e:
            logging.error(f"Error downloading {song} by {artist}: {e}")
```

```python
from pydub import AudioSegment
import os

# Function to extract chorus from audio file
def extract_chorus(file_path):
    # Load the audio file
    audio = AudioSegment.from_file(file_path)

    # Set the duration of each segment in milliseconds (adjust as needed)
    segment_duration = 10 * 1000  # 10 seconds

    # Iterate over the audio in segments
    for i in range(0, len(audio), segment_duration):
        # Extract a segment
        segment = audio[i:i+segment_duration]

        # For demonstration purposes, let's assume every segment is a chorus
        is_chorus = True

        # If the segment is determined to be the chorus, you can save it or perform further processing
        if is_chorus:
            output_file_path = os.path.splitext(file_path)[0] + "_chorus.mp3"
            segment.export(output_file_path, format="mp3")
            print(f"Chorus extracted successfully and saved to {output_file_path}")

# Directory where downloaded songs are stored
download_directory = "downloaded_songs"

# Iterate over all downloaded songs in the directory
for file_name in os.listdir(download_directory):
    if file_name.endswith(".mp3"):
        file_path = os.path.join(download_directory, file_name)
        extract_chorus(file_path)
```

```python
import librosa

# Function to extract audio features using Librosa
def extract_audio_features(file_path):
    # Load the audio file
    y, sr = librosa.load(file_path)

    # Extract audio features
    tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rmse = librosa.feature.rms(y=y)
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zero_crossing_rate = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)

    # Print extracted audio features
    print("Tempo:", tempo)
    print("Beats:", beats)
    print("Chroma STFT:", chroma_stft.shape)  # Print shape since it's a 2D matrix
    print("RMS Energy:", rmse)
    print("Spectral Centroid:", spectral_centroid)
    print("Spectral Bandwidth:", spectral_bandwidth)
    print("Spectral Rolloff:", spectral_rolloff)
    print("Zero Crossing Rate:", zero_crossing_rate)
    print("MFCCs:", mfcc.shape)  # Print shape since it's a 2D matrix

# Directory where extracted chorus files are stored
chorus_directory = "/content/downloaded_songs"

# Iterate over all extracted chorus files in the directory
for file_name in os.listdir(chorus_directory):
    if file_name.endswith(".mp3"):
        file_path = os.path.join(chorus_directory, file_name)
        print(f"Extracting audio features for: {file_name}")
        extract_audio_features(file_path)
        print("\n")  # Add newline for better readability
```

```python
import billboard

def extract_songs_from_chart(chart_name, chart_date=None):

    # Fetch the Billboard chart
    chart = billboard.ChartData(chart_name, date=chart_date)

    # Initialize an empty list to store song information
    songs = []

    # Iterate over chart entries and extract song details
    for entry in chart:
        song_info = {
            'title': entry.title,
            'artist': entry.artist,
            'peak_position': entry.peakPos,
            'last_position': entry.lastPos,
            'weeks_on_chart': entry.weeks,
            'current_rank': entry.rank,
            'is_new': entry.isNew
        }
        songs.append(song_info)

    return songs

# Example usage
if __name__ == "__main__":
    chart_name = 'hot-100'  # Example: Hot 100 chart
    chart_date = None  # Leave as None for the latest chart, or specify a date in YYYY-MM-DD format
    songs = extract_songs_from_chart(chart_name, chart_date)
    for song in songs:
        print(song)
```

```python
import billboard
from tabulate import tabulate

def extract_songs_from_chart(chart_name, chart_date=None):
    """
    Extracts songs from the specified Billboard chart.

    Args:
    chart_name (str): The name of the Billboard chart, e.g., 'hot-100' or 'pop-songs'.
    chart_date (str, optional): The date of the chart in YYYY-MM-DD format. Defaults to None (latest chart).

    Returns:
    list: A list of dictionaries containing song information.
    """
    # Fetch the Billboard chart
    chart = billboard.ChartData(chart_name, date=chart_date)

    # Initialize an empty list to store song information
    songs = []

    # Iterate over chart entries and extract song details
    for entry in chart:
        song_info = {
            'Title': entry.title,
            'Artist': entry.artist,
            'Peak Position': entry.peakPos,
            'Last Position': entry.lastPos,
            'Weeks on Chart': entry.weeks,
            'Current Rank': entry.rank,
            'Is New': entry.isNew
        }
        songs.append(song_info)

    return songs

# Example usage
if __name__ == "__main__":
    chart_name = 'hot-100'  # Example: Hot 100 chart
    chart_date = None  # Leave as None for the latest chart, or specify a date in YYYY-MM-DD format
    songs = extract_songs_from_chart(chart_name, chart_date)

    # Display songs in a table
    print(tabulate(songs, headers="keys", tablefmt="grid"))
```

```python
import csv
def save_to_csv(data, filename):

    # Get the keys (column names) from the first dictionary in the list
    fieldnames = data[0].keys()

    # Write data to CSV file
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

# Example usage
if __name__ == "__main__":
    chart_name = 'hot-100'
    chart_date = None  # Leave as None for the latest chart, or specify a date in YYYY-MM-DD format
    songs = extract_songs_from_chart(chart_name, chart_date)

    # Save songs to a CSV file
    save_to_csv(songs, 'billboard_songs.csv')
```
